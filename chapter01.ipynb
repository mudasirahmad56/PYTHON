##### Escape sequence

```python
print("mudasir \nmubashir\nmusawir") # New line
print("mudasir \tmubashir\tmusawir") # Tab space
print("mudasir \\ mubashir \\ musawir") # Backslash
print("mudasir \' mubashir \' musawir") # Single quote
print("mudasir \" mubashir \" musawir") # Double quote
print("mudasir \b mubashir \b musawir") # Backspace
print("mudasir \r mubashir \r musawir") # Carriage return
print("mudasir \f mubashir \f musawir") # Form feed
```

### conditions

```python
for i in range(1, 6):
    print(i)

# example loop with a break to avoid an actual infinite loop
count = 0
while True:
    print("loop with break (runs 3 times)")
    count += 1
    if count >= 3:
        break

# simulate a do-while loop in Python: execute body once, then check condition
count2 = 0
while True:
    print("do while loop")
    count2 += 1
    # replace the condition below with the real exit condition as needed
    if count2 >= 1:
        break

## break and continue statements
for fruit in ["apple", "banana", "cherry" , "date", "elderberry"]:
    if fruit == "banana":
        break
    print(fruit)

for fruit in ["apple", "banana", "cherry" , "date", "elderberry"]:
    if fruit == "banana":
        continue
    print(fruit)
```
