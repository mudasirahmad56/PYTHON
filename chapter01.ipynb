### varibales

```python
# Dynamic typing vs Static typing
# Dynamic binding vs Static binding
# stylish diclaration techniques
```

```python
#dynamic typing
x = 10
print(x)

#static typeing (using in other languages like C, Java)
int y = 20

```

```python
#dynamic binding (you can change the type of variable in python in the same program)
x = "hello"
print(x)
x = 'mudasir'
print(x)

#static binding (using in other languages like C, Java you cannot change the type of variable in the same program)
int z = 30
```

```python
a = 5
b = 10
c = 5
print(a,b,c)

a,b,c = 5,10,5
print(a,b,c)

```
