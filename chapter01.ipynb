### operators in python

```python
# arithmetic operations
# Relational operators
# Logical operators
# Bitwise operators
# assignment operators
# membership operators
```

```python
# Arithmetic operators
print(5 + 2)    # addition
print(5 - 2)    # subtraction
print(5 * 2)    # multiplication
print(5 / 2)    # division
print(5 // 2)   # floor division
print(5 % 2)    # modulus
print(5 ** 2)   # exponentiation
```

```python
# Relational operators
print(5 > 2)    # greater than
print(5 < 2)    # less than
print(5 >= 5)   # greater than or equal to
print(5 <= 5)   # less than or equal to
print(5 == 5)   # equal to
print(5 != 5)   # not equal to
```

```python
# Logical operators
print(1 and 0)    # logical AND
print(1 or 0)     # logical OR
print(not 1)      # logical NOT
```

```python
# bitwise operators
print(5 & 3)    # bitwise AND
print(5 | 3)    # bitwise OR
print(5 ^ 3)    # bitwise XOR
print(~5)       # bitwise NOT
print(5 << 1)  # left shift
print(5 >> 1)  # right shift
```

```python
# Assignment operators
# =
# a = 5

a = 5
# a = a + 5
a += 5  # equivalent to a = a + 5
print(a)
```

```python
# membership operators
print('a' in 'apple')

print(1 in [2, 3, 4, 5])
```

```python
# menu driven calculator

fnum = int(input("Enter first number: "))
snum = int(input("Enter second number: "))

op = input("Enter operator (+, -, *, /): ")
if op == '+':
    print("Result:", fnum + snum)
elif op == '-':
    print("Result:", fnum - snum)
elif op == '*':
    print("Result:", fnum * snum)
elif op == '/':
    if snum != 0:
        print("Result:", fnum / snum)
    else:
        print("Error: Division by zero")
else:
    print("Invalid operator")
```
