##### Escape sequence

```python
print("mudasir \nmubashir\nmusawir") # New line
print("mudasir \tmubashir\tmusawir") # Tab space
print("mudasir \\ mubashir \\ musawir") # Backslash
print("mudasir \' mubashir \' musawir") # Single quote
print("mudasir \" mubashir \" musawir") # Double quote
print("mudasir \b mubashir \b musawir") # Backspace
print("mudasir \r mubashir \r musawir") # Carriage return
print("mudasir \f mubashir \f musawir") # Form feed
```

#### intergers 

```python
print (2+3) # Addition
print (5-2) # Subtraction
print (4*3) # Multiplication
print (8/2) # Division
print (7//3) # Floor Division
print (5%2) # Modulus
print (2**3) # Exponentiation
print (10%3) # Modulus
print (10//3) # Floor Division
print (2**0.5) #square root
print (round(2**0.5, 4)) # Rounding
```
